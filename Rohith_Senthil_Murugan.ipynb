```python
import pandas as p
import numpy as n
import sklearn
```

```python
from sklearn.datasets import load_boston
ob=load_boston()
```

```python
ob.keys()
```

```python
bos=p.DataFrame(ob.data,columns=ob['feature_names'])
bos.head()
```

```python
bos["TARG"]=ob.target
bos.head()
```

```python
bos.isnull()
```

```python
bos.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split as tsp
Q=bos.drop('TARG',axis=1)
R=bos['TARG']
Q_train, Q_test, R_train, R_test =tsp(Q,R,test_size=0.15,random_state=5)

print(Q_train.shape)
print(Q_test.shape)
print(R_train.shape)
print(R_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as msr

#Fitting the data
bos_model=LinearRegression()
bos_model.fit(Q_train,R_train)
```

```python
R_train_predict=bos_model.predict(Q_train)
R_train_rms=n.sqrt(msr(R_train,R_train_predict))
print('Performace of the model for the training set\nThe Root Mean Sqaured Error is {}'.format(R_train_rms))

R_test_predict=bos_model.predict(Q_test)
R_test_rms=n.sqrt(msr(R_test,R_test_predict))
print('\nPerformace of the model for the testing set\nThe Root Mean Sqaured Error is {}'.format(R_test_rms))
```
